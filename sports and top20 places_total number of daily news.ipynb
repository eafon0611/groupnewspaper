```python
import pandas as pd
data = pd.DataFrame(pd.read_table("Dundee1950.tsv"))
data['place']=[string.title() for string in data['place']]
data=data.dropna()# delete Ｎan
data['title']=[string.title() for string in data['title']]
data=data.drop_duplicates(subset=['title','date','place'], keep=False)
data
```

```python
from collections import Counter
lst= data['title']
c = Counter(i.capitalize() for x in lst for i in x.split())
print(c.most_common(200))
```

```python
rank=data.copy(deep=True)
rank['freq'] = rank.groupby(data['place'])['place'].transform('count')
rank=rank.sort_values("freq",ascending=False)
rank=rank.loc[rank['freq']>50]
rank
```

```python
rank2=rank.drop_duplicates(subset='place', keep='first', inplace=False)
rank2.reset_index(drop=True, inplace=True)
rank2
```

```python
rank=rank.loc[(rank['freq']>50) & (rank['freq']<2439)] #去除鄧迪的數據保留前五十
rank=rank.sort_values(["date"],ascending=True)
rank.head()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

# 照日期排，每天運動新聞佔的比例   4/14,8/7 全天都是運動新聞，我是有沒有看錯？
f, ax = plt.subplots(figsize=(6,35))
sns.set_color_codes("pastel")
sns.barplot(x="daily total", y="date",data=rankdaily,color="b",label="Rank 50")
sns.set_color_codes("muted")
sns.barplot(x="daily total", y="date",data=sportdaily,color="b",label="Sport")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="2015",xlabel="Total Number of Daily News")
sns.despine(left=True, bottom=True)
```

```python
# 照運動新聞多到少排列，和每天運動新聞佔的比例, 4/14,8/7 照道理說是整個深藍，可是這樣就看不到了？
sportdaily=sportdaily.sort_values(["daily total"],ascending=False)
rankdaily=rankdaily.sort_values(["daily total"],ascending=False)
f, ax = plt.subplots(figsize=(6,35))
sns.set_color_codes("pastel")
sns.barplot(x="daily total", y="date",data=rankdaily,color="b",label="Rank 50")
sns.set_color_codes("muted")
sns.barplot(x="daily total", y="date",data=sportdaily,color="b",label="Sport")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="2015",xlabel="Total Number of Daily News")
sns.despine(left=True, bottom=True)
```
